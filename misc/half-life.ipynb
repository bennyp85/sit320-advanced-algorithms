```python
import numpy as np
import matplotlib.pyplot as plt

# Correcting the data type issue
import pandas as pd
import numpy as np
half_life = 33
dose_20mg = 20
dose_10mg = 10
time_points = np.arange(0, 200, 1)
decay_constant = np.log(2) / half_life
# calculate the decay of the 20mg dose with new dose taken every 24 hours
conc_20mg = dose_20mg * np.exp(-decay_constant * time_points)
conc_10mg = dose_10mg * np.exp(-decay_constant * time_points)


df = pd.DataFrame({'Time (hours)': time_points, 'Concentration_20mg (mg)': conc_20mg, 'Concentration_10mg (mg)': conc_10mg})
df.head(48)
df.plot(x='Time (hours)', y=['Concentration_20mg (mg)', 'Concentration_10mg (mg)'], figsize=(10, 5))
```
