## Board Definition

```python
#define the board
board={1:' ', 2:' ', 3:' ',
       4:' ', 5:' ', 6:' ',
       7:' ', 8:' ', 9:' '}
```

```python
#function to print the board
def printBoard(board):
    print(board[1]+'|'+board[2]+'|'+board[3])
    print('-+-+-')
    print(board[4] + '|' + board[5] + '|' + board[6])
    print('-+-+-')
    print(board[7] + '|' + board[8] + '|' + board[9])
    print('\n')
```

```python
printBoard(board)
```

```python
#function to check is a certain position in the board is empty.
def spaceIsFree(position):
    
    if (board[position]==' '):
        return True
    else:
        return False
```

```python
#method to insert letter in space
def insertLetter(letter, position):
    
    if (spaceIsFree(position)):
        board[position]=letter
        printBoard(board)
        
        if (chkDraw()):
            print('Draw!')
        elif (chkForWin()):
            if (letter=='X'):
                print('Bot wins!')
            else:
                print('You win!')
        return
    
    else:
        print('Position taken, please pick a different position.')
        position=int(input('Enter new position: '))
        insertLetter(letter, position)
        
        return
```

```python
#function to check if board is draw
def chkDraw():
    
    for key in board.keys():
        if (board[key]==' '):
            return False
    return True

#function to check if one user has won
def chkForWin():
    
    if (board[1]==board[2] and board[1]==board[3] and board[1] !=' '):
        return True
    elif (board[4]==board[5] and board[4]==board[6] and board[4] !=' '):
        return True
    elif (board[7]==board[8] and board[7]==board[9] and board[7] !=' '):
        return True
    elif (board[1]==board[5] and board[1]==board[9] and board[1] !=' '):
        return True
    elif (board[3]==board[5] and board[3]==board[7] and board[3] !=' '):
        return True
    elif (board[1]==board[4] and board[1]==board[7] and board[1] !=' '):
        return True
    elif (board[2]==board[5] and board[2]==board[8] and board[2] !=' '):
        return True
    elif (board[3]==board[6] and board[3]==board[9] and board[3] !=' '):
        return True
    else:
        return False
```

```python
#function to check who won
def chkMarkForWin(mark):
    
    if (board[1]==board[2] and board[1]==board[3] and board[1] ==mark):
        return True
    elif (board[4]==board[5] and board[4]==board[6] and board[4] ==mark):
        return True
    elif (board[7]==board[8] and board[7]==board[9] and board[7] ==mark):
        return True
    elif (board[1]==board[5] and board[1]==board[9] and board[1] ==mark):
        return True
    elif (board[3]==board[5] and board[3]==board[7] and board[3] ==mark):
        return True
    elif (board[1]==board[4] and board[1]==board[7] and board[1] ==mark):
        return True
    elif (board[2]==board[5] and board[2]==board[8] and board[2] ==mark):
        return True
    elif (board[3]==board[6] and board[3]==board[9] and board[3] ==mark):
        return True
    else:
        return False
```

```python
player = 'O'
bot ='X'
```

```python
#function for player move
def playerMove():
    position=int(input('Enter position for O:'))
    insertLetter(player, position)
    return

#function for bot move
def compMove():
    
    bestScore=-1000 #lowest to start with, for comaparison to the score after a certain move
    
    bestMove=0
    
    for key in board.keys(): #for loop to find emplty slots, make a move, calculate the score and see if its higher then best score
        if (board[key]==' '):
            board[key]=bot
            score = minimax(board, 0, False) #calculate the score
            board[key] = ' ' #set board back to what it was
            if (score > bestScore): #compare the score from a certain move to the best score.
                bestScore = score
                bestMove = key
                
    #at the end of the loop, we have the best move figured out.
    insertLetter(bot, bestMove)
    return
```

```python
def minimax(board, depth, isMaximizing):
    
    if (chkMarkForWin(bot)):
        return 1
    elif (chkMarkForWin(player)):
        return -1
    elif (chkDraw()):
        return 0
    
    if isMaximizing:
        
        bestScore = -1000
        
        for key in board.keys():
            if board[key]==' ':
                board[key]=bot
                score = minimax(board, 0, False)
                board[key]= ' '
                if (score>bestScore):
                    bestScore = score
        return bestScore
    
    else:
        
        bestScore = 1000
        
        for key in board.keys():
            if board[key]==' ':
                board[key]=player
                score = minimax(board, 0, True)
                board[key]= ' '
                if (score<bestScore):
                    bestScore = score
        return bestScore
    
```

```python
while not chkForWin():
    compMove()
    playerMove()
```

```python
from unittest.mock import patch

def test_insertLetter():
    print('Running test_insertLetter...')
    
    # Reset the board
    for i in range(1, 10):
        board[i] = ' '
    
    # Case where position is free
    insertLetter('X', 1)
    assert board[1] == 'X', "Test case failed! Expected 'X' at position 1."
    
    # Case where position is not free. Mock the input function to return '2'
    with patch('builtins.input', return_value='2'):
        insertLetter('O', 1)
    assert board[2] == 'O', "Test case failed! Expected 'O' at position 2, not at position 1."
    
    print('test_insertLetter passed!\n')


def test_chkDraw():
    print('Running test_chkDraw...')
    
    # Reset the board
    for i in range(1, 10):
        board[i] = ' '
    
    # Not a draw case
    assert not chkDraw(), "Test case failed! Expected False as board is not full."

    # Draw case
    for i in range(1, 10):
        board[i] = 'X'
    assert chkDraw(), "Test case failed! Expected True as board is full."

    print('test_chkDraw passed!\n')


def test_chkForWin():
    print('Running test_chkForWin...')
    
    # Reset the board
    for i in range(1, 10):
        board[i] = ' '
    
    # Case where no one has won
    assert chkForWin() == False, 'Test case failed! Expected False when no one has won.'
    
    # Case where there is a horizontal win
    for i in range(1, 4):
        board[i] = 'X'
    assert chkForWin() == True, 'Test case failed! Expected True when there is a horizontal win.'
    
    # Case where there is a vertical win
    for i in range(1, 4):
        board[i] = ' '
    for i in range(1, 8, 3):
        board[i] = 'O'
    assert chkForWin() == True, 'Test case failed! Expected True when there is a vertical win.'
    
    # Case where there is a diagonal win
    for i in range(1, 8, 3):
        board[i] = ' '
    for i in range(1, 10, 4):
        board[i] = 'X'
    assert chkForWin() == True, 'Test case failed! Expected True when there is a diagonal win.'
    
    print('test_chkForWin passed!\n')

def test_spaceIsFree():
    print('Running test_spaceIsFree...')
    
    # Position is free
    board[1] = ' '
    assert spaceIsFree(1), "Test case failed! spaceIsFree returned False, expected True."

    # Position is not free
    board[1] = 'X'
    assert not spaceIsFree(1), "Test case failed! spaceIsFree returned True, expected False."

    print('test_spaceIsFree passed!\n')

def test_chkMarkForWin():
    print('Running test_chkMarkForWin...')
    
    # Reset the board
    for i in range(1, 10):
        board[i] = ' '
    
    # Winning case
    board[1] = 'O'
    board[2] = 'O'
    board[3] = 'O'
    assert chkMarkForWin('O'), "Test case failed! Expected True as 'O' has a winning line."

    # Losing case
    board[3] = 'X'
    assert not chkMarkForWin('O'), "Test case failed! Expected False as 'O' does not have a winning line."

    print('test_chkMarkForWin passed!\n')

# Run the tests
test_spaceIsFree()
test_insertLetter()
test_chkDraw()
test_chkForWin()
test_chkMarkForWin()




```
